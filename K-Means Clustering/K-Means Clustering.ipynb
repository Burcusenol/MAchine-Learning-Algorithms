```python
X=pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_X=scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=2)
```

```python
cluster_labels=model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
X["Cluster"]=cluster_labels
```

```python
X
```

```python
X.corr()
```

```python
X.corr()["Cluster"].iloc[:-1]
```

```python
plt.figure(figsize=(12,6),dpi=200)
X.corr()["Cluster"].iloc[:-1].sort_values().plot(kind="bar")
```

```python
ssd=[]

for k in range(2,10):
    model=KMeans(n_clusters=k)
    model.fit(scaled_X)


    ssd.append(model.inertia_) #sum squared distince point -->cluster center
    
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,'o--')
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()
```

# K-Means Color Quantization

```python
import matplotlib.image as mpimg   #.jpg .png ---->numpy array
```

```python
image_as_array=mpimg.imread("./palm_trees.jpg")
```

```python
image_as_array
```

```python
image_as_array.shape  # (H,W,C)
```

```python
plt.figure(dpi=200)
plt.imshow(image_as_array)
```

```python
#(Height,Witdh,Color) ---> 2D (H*W,C)
(h,w,c)=image_as_array.shape
```

```python
h
```

```python
image_as_array2d=image_as_array.reshape(h*w,c)
```

```python
image_as_array2d
```

```python
len(image_as_array2d.shape)
```

```python
len(image_as_array.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=6)
```

```python
labels=model.fit_predict(image_as_array2d)
```

```python
1401*934
```

```python
labels
```

```python
rgb_codes=model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
labels
```

```python
quantized_image=np.reshape(rgb_codes[labels],(h,w,c))
```

```python
plt.imshow(quantized_image)
```
